```python
df.iloc[:,3:9]
```

```python
seen = df.iloc[:,3:9].copy()
```

```python
seen.columns = ["seen_episode1","seen_episode2",
                "seen_episode3","seen_episode4",
                "seen_episode5","seen_episode6"]
```

```python
seen
```

```python
seen = seen.notnull()
```

```python
seen
```

```python
df.iloc[:,9:15]
```

```python
ranking = df.iloc[:,9:15].copy()
```

```python
ranking
```

```python
ranking.columns
```

```python
ranking.columns = ["ranking_"+str(i) for i in range(1,7)] # list comprehension
```

```python
ranking
```

```python
ranking.info()
```

```python
ranking.astype(float)
```

```python
ranking = ranking.astype(float)
```

```python
pd.concat([clean,seen, ranking])    #combine three data on which we transform before 
```

```python
final = clean.merge(seen,left_index=True, right_index=True)
```

```python
final = final.merge(ranking,left_index=True, right_index=True)
```
